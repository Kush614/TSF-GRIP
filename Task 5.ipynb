```python

plt.figure(figsize=(20, 15))
plt.subplot(2,2,1)
sns.scatterplot(x='Ship Mode',y = 'Sales', data = df)
plt.subplot(2,2,2)
sns.scatterplot(x='Segment',y = 'Sales', data = df)
plt.subplot(2,2,3)
sns.scatterplot(x='Category',y = 'Sales', data = df)
plt.subplot(2,2,4)
sns.scatterplot(x='Region',y = 'Sales', data = df)
plt.show()
```

As we can see there are lot of outliers in the sales feature but we can't ignore these outliers.

- **Standard Class** seems to bring in the most sales.
- **Consumer** also have the high sales figures in segment.
- **technology and office supplies** have the almost same high sales.
- All the regions seems to have similar sales with some regions like **Central** having some outliers in sales.

```python
plt.figure(figsize=(20, 15))
plt.subplot(2,2,1)
sns.scatterplot(x='Ship Mode',y = 'Profit', data = df)
plt.subplot(2,2,2)
sns.scatterplot(x='Segment',y = 'Profit', data = df)
plt.subplot(2,2,3)
sns.scatterplot(x='Category',y = 'Profit', data = df)
plt.subplot(2,2,4)
sns.scatterplot(x='Region',y = 'Profit', data = df)
plt.show()
```

- **Standard Class** has shown huge losses but profits as well
- **Consumer** segment has also shown huge losses but profits as well
- **Office supplies** category has shown both porfits and loss

```python
plt.figure(figsize=(20, 15))
plt.subplot(2,2,1)
sns.barplot(x = 'Ship Mode', y = 'Sales', data = df)
plt.subplot(2,2,2)
sns.barplot(x = 'Segment', y = 'Sales', data = df)
plt.subplot(2,2,3)
sns.barplot(x = 'Category', y = 'Sales', data = df)
plt.subplot(2,2,4)
sns.barplot(x = 'Region', y = 'Sales', data = df)
plt.show()
```

The bar plots shows the proper results for sales as outliers doesn't have much effect here and we have a much clearer picture.

- All shipping modes have almost the same sales with **Second Class and Same Day** edging a little higher from the other modes.
- **Home Office** has higher sales than the other segments.
- **Technology** category has higher sales than others.
- **Southern and Eastern** region have higher sales than other regions.

```python
plt.figure(figsize=(20, 15))
plt.subplot(2,2,1)
sns.violinplot(x = 'Ship Mode', y = 'Profit', data = df)
plt.subplot(2,2,2)
sns.violinplot(x = 'Segment', y = 'Profit', data = df)
plt.subplot(2,2,3)
sns.violinplot(x = 'Category', y = 'Profit', data = df)
plt.subplot(2,2,4)
sns.violinplot(x = 'Region', y = 'Profit', data = df)
plt.show()
```

- **Standard Class** shows high profit but loss as well, the manager can work on reducing the loss so as to maximize overall profit.
- **Consumer** segment shows both high profit and loss. Manager can work on reducing the loss to maximize profit.
- **Technology** category shows both high profit and loss, loss should be decreased.
